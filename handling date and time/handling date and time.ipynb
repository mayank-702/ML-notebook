```python
#converting to date and time datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#extract year
date['date_year'] = date['date'].dt.year
```

```python
date.head()
```

```python
# extract month
date['date_month'] = date['date'].dt.month
```

```python
date.head()
```

```python
#month name
date['date_monthname'] = date['date'].dt.month_name()
```

```python
date.head()
```

```python
# extract days

date['date_days'] = date['date'].dt.day
```

```python
date.head()
```

```python
# extract day of week 

date['date_daysofweek'] = date['date'].dt.dayofweek
date.head()
```

```python
# day of week name

date['date_weekname'] = date['date'].dt.day_name()
```

```python
date.head()
```

```python
# is weekend
date['date_is_weekend'] = np.where(date['date_weekname'].isin(['Sunday','Saturday']),1,0)
date.drop(columns = ['product_id','city_id','orders']).head()
```

```python
#extract week of the year
# extract week of the year
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

```python
# extract quarter
date['date_quarter'] = date['date'].dt.quarter
date.head()
```

```python
#extract semester
date['semester'] = np.where(date['date_quarter'].isin([1,2]),1,2)
date.head()
```

```python
# extract time elapsed between dates
import datetime
today = datetime.datetime.today()
```

```python
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
#months passed
date['months_passed'] = np.round((today - date['date']) / pd.Timedelta(1, 'm'),0)
date.head()
```
